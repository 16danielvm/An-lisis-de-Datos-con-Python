Grafico Lineal

```python
#Importar librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Carga el dataset 'vehiculos.csv' como 'tabla'
tabla = pd.read_csv('vehiculos.csv')
```

```python
#Muestra los primeros 10 datos de 'tabla'
tabla.head(10)
```

```python
#Grafica la cantidad de valores de la columna 'vehicle_class' de tipo linea y color rojo
tabla.vehicle_class.value_counts().plot(kind='line', color='red') 
```

```python
#Grafica de barras de la cantidad de valores en la columna 'vehicle class', en color naranja
tabla.vehicle_class.value_counts().plot(kind='bar', color='orange') 
```

```python
#Grafica en barras horizontales de la cantidad de valores de la columna 'vehicle_class'
tabla.vehicle_class.value_counts().plot(kind='barh') 
```

```python
#Grafico pastel de la cantidad de valores de la columna 'vehicle_class'
tabla.vehicle_class.value_counts().plot(kind='pie', colormap='Pastel1') 
```

```python
#Scatterplot de 'scaled_variance_minor_axis' vs 'skewness_major_axis' de tamao (10,5)
tabla.plot(kind='scatter', x='scaled_variance_minor_axis', y='skewness_major_axis', figsize=(10,5)) 
```
